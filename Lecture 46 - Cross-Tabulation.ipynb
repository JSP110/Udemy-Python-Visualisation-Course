```python
import pandas as pd
```

```python
from io import StringIO
```

```python
data="""
Sample Animal Intelligence
1 Dog Smart
2 Dog Smart
3 Cat Dumb
4 Cat Dumb
5 Dog Dumb
6 Cat Smart"""
```

```python
dframe = pd.read_table(StringIO(data),sep='\s+')
```

```python
dframe
```

```python
pd.crosstab(dframe.Animal,dframe.Intelligence,margins=True)
```

```python
pd.crosstab(dframe.Animal,dframe.Intelligence,margins=False)
```
